```python
def get_tail(head, map):
    """moves mapward until an already-visited node is encountered
       returns path. to detect already visited nodes a global set 'seen'
       is referenced and modified"""
    seen.add(head)   # mark as read
    path = [head]    # initialize the path
    next = map[head] # move down
    
    while not next in seen:  # footle on down
        seen.add(next)       # mark as read
        path.append(next)    # keep
        next = map[next]    # apply the mapping
        
        
    tail = next  # the déja vu node
    
    return path, tail
```
